# Matplotlib 数据可视化

```python
import matplotlib as mpl
```

```python
import matplotlib.pyplot as plt        #一般只用plt
```

```python
import numpy as np
x = np.linspace(0 ,10, 101)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
y2 = np.cos(x)
```

```python
plt.plot(x, y)
plt.plot(x, y2)
```

```python
plt.plot(x, y)
plt.plot(x, y2, '--',color = 'red')
plt.xlim(-5,15)
plt.ylim(0,1.5)
plt.show()
```

```python
plt.plot(x, y, label = 'sin(x)')
plt.plot(x, y2, '--',color = 'red',label = 'cos(x)')
plt.axis([-1,12,-2,2])
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()               #图例
plt.title('sinx and cosx')        #标题
plt.show()
```

# Scatter Plot 

```python
plt.scatter(x, y)
plt.scatter(x, y2, color = 'red')
plt.show()
```
